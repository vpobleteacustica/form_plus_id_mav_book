### Describir laboratorios, bibliotecas y otras instalaciones que utilizan los profesores y estudiantes del programa, con su respectivo equipamiento

<table align="left" style="width:100%">
    <tbody style="font-size: 14px";>
        <tr>
            <td style="width: 7px;background:#036ED1;"></td>
            <td style="outline: thin solid" rowspan="1" colspan="1"><div style="text-align: justify; ">Los estudiantes y cuerpo académico del Magíster en Acústica y Vibraciones cuentan con las siguientes instalaciones para la realización de sus actividades:</div><br>
            <div style="text-align: justify; "><strong><em>Laboratorios</em></strong></div><br>
            <div style="text-align: justify; "><strong>Laboratorios Integrados de la Facultad de Ciencias de la Ingeniería:</strong> En estos laboratorios se encuentra el Laboratorio de Acústica y Vibraciones con moderno equipamiento tecnológico, por ejemplo, demostradores completos para el estudio de ondas,  generadores/osciladores mecánicos de ondas, un sistema avanzado de cubeta de ondas, generadores de funciones, sintetizador de Fourier, un set de láminas Chladni para el estudio de ondas, tubos de resonancia para ondas acústicas, Osciloscopio digital, shakers excitadores de vibración, Luz estroboscópica, Acelerómetros, Martillo de impactos, Transductores de fuerza, Calibradores, Sonda medidora de Intensidad Acústica, Cabezas de impedancia, Vibrómetro Láser, Micrófonos de Medición y preamplificadores de &#189;, &#188; y 1″, Audiómetro, oído artificial, más varios sistemas de adquisición y análisis de ruido y vibración.</div><br> 
            <div style="text-align: justify; "><strong>Laboratorio de Modelaciones Acústicas:</strong> Laboratorio de uso exclusivo del Instituto de Acústica habilitado para apoyar las actividades de aprendizaje y desarrollo de software de los estudiantes. El laboratorio cuenta con computadores, hardware (sistemas de adquisición de datos de alta resolución) y software licenciado (Matlab, ToolBox Signal Analisis, ToolBox Data Acquisition, CADNA A, Predictor–Lima, Insul, Ease) adecuados al desarrollo de programas aplicados al campo de la acústica.</div><br>
            <div style="text-align: justify; "><strong>Laboratorio de Absorción y Resistencia de Flujo: </strong> Este laboratorio cuenta con un sistema de medición de Coeficiente de Reflexión Complejo e Impedancia Acústica Específica a incidencia normal, basado en el método de la Función de Transferencia (ISO 10534-2, Parte 2). La adquisición de señales se realiza a través de un sistema Análogo-Digital de 4 canales Photon+ marca Bruel &#38; Kjaer y las señales son procesadas digitalmente por medio de software. Este sistema permite obtener las características de absorción, reflexión, resistencia al flujo e impedancia de diferentes materiales y configuraciones de éstos, información de gran utilidad para el desarrollo de proyectos y soluciones en el área. Además, se cuenta con un sistema de medición de resistencia de flujo basado en la norma ASTM C522-87, un medidor de porosidad y un micrómetro digital. La resistencia al flujo y la porosidad de los materiales está íntimamente ligada a las propiedades acústicas de éstos, especialmente materiales porosos y fibrosos para absorción del sonido.</div><br>
            <div style="text-align: justify; "><strong>Laboratorio de Caracterización de Materiales: </strong> El laboratorio (del Instituto de Materiales y Procesos Termomecánicos) se enfoca en el estudio de las estructuras fundamentales de diversos materiales a través del uso de difracción de Rayos X (DRX) y Microscopia Raman confocal acoplada con Microscopia de Fuerza Atómica (AFM). Su utilización es para para investigación y control de calidad en minerales, polímeros y elementos orgánicos. El DRX determina las propiedades cristalográficas de los materiales, permitiendo la identificación de fases en sólidos en estado cristalino. El Raman permite el estudio de los modos vibracionales, rotaciones y otros de baja frecuencia en un sistema. El AFM permite la caracterización de superficie a nanoescala con alta resolución.</div><br>
           <div style="text-align: justify; "><strong>Laboratorio Fondequip:</strong> El año 2015, el Instituto de Acústica de la Universidad Austral postuló al IV Concurso de Equipamiento Científico y Tecnológico Mediano FONDEQUIP de CONICYT. Participó con el proyecto Desarrollo de aplicaciones en acústica mediante la técnica de arreglo de micrófonos de alta definición e imágenes acústicas, adjudicándose la compra de un sistema de arreglo de micrófonos (Beamforming System) o Cámara Acústica, marca Bruel &#38; Kjaer. Este equipo permite visualizar el comportamiento del sonido, y se basa en métodos de procesamiento digital de señales, permitiendo extraer e integrar la información acústica que proviene de múltiples fuentes sonoras. El sistema de hardware y software (al igual que los micrófonos) son modulares, y presenta versatilidad en su configuración, sin perder calidad de los datos, y abriendo posibilidades a otras aplicaciones y futuras ampliaciones. Incluye arreglo pentangular, pizza (o redondo) y esférico. Además, cuenta con el Software Reflex, en versión full, con licencia educacional por cinco años.</div><br>
           <div style="text-align: justify; "><strong>Cámaras reverberantes:</strong> Se cuenta con un sistema de dos cámaras reverberantes adyacentes, una de las cuales se emplea para realizar ensayos de medición de absorción sonora y potencia sonora. Las dos salas se pueden usar simultáneamente para realizar ensayos de aislamiento acústico de divisiones (paredes) y elementos constructivos (puertas, ventanas, etc.). Además, un sistema de losa normalizada permite ensayar ruido de impacto sobre una losa patrón, e incluso ensayos de nuevos sistemas constructivos para aislar el ruido de impacto en edificios. Todas las salas cumplen con las exigencias de las normativas ISO para la realización de ensayos, en cuanto a volumen, dimensiones y demás condiciones acústicas relevantes. Para realizar estos estudios, el Instituto cuenta con una máquina de impactos normalizada, fuente dodecaédrica y sonómetros de última generación.</div><br>
           <div style="text-align: justify; "><strong>Salas anecoicas:</strong> Situadas en el Instituto de Acústica, el MAV dispone para sus actividades de dos cámaras anecoicas, de 26,32 m3 y 16,8 m3 útiles, completamente equipadas (micrófonos de medición, sonómetros y analizadores, amplificadores, sistemas de desplazamiento controlados por computador, sistemas de adquisición de datos de alta resolución, etc). Estas instalaciones permiten realizar distintos tipos de mediciones (directividad de parlantes, directividad de micrófonos, radiación de instrumentos musicales, potencia acústica, efectos de difracción del sonido, comportamiento de resonadores acústicos, respuesta de frecuencia, etc), tanto para investigación como para las actividades de aprendizaje de los estudiantes.</div><br>
           <div style="text-align: justify; "><strong>Laboratorio de Audio y Video:</strong> El laboratorio cuenta con computadores e interfaces que permiten a los estudiantes editar producciones musicales, así como realizar post-producción de audio para video y edición de video. Dado que actualmente existe software y hardware de edición basados tanto en sistema Mac como PC, este laboratorio cuenta con ambas plataformas, de esta manera el estudiante tiene la posibilidad de conocer y practicar con ambos sistemas, logrando acumular una mejor y mayor experiencia en el tema.</div><br>
           <div style="text-align: justify; "><strong>Estudio de Grabación y Post-producción:</strong> Los estudiantes que realizan trabajos en área del sonido, pueden acceder al estudio de grabación y post producción, que cuenta con todo el equipamiento electroacústico necesario para realizar grabaciones, edición, masterización, post-producción y audio para video, de acuerdo a los más altos estándares de calidad. Sus instalaciones cuentan con aproximadamente 160 m2 distribuidos en un hall, dos salas de control, estudio y sala de edición de audio para video. La infraestructura e instrumental principal consta de microfonía (Neumann U87AI/M147/TLM49 Set, Earthworks DK50/R, Earthworks SR30mp, Earthworks QTC50mp, Earthworks M30BX, Sony C-48, AKG414BXLS, ShureKSM44, AudioTechnica 4050/4051/4053, Sennheiser MD421II, Rode NTK, Rode NT2A, Rode NT4, Shure Beta 58/56/52, Shure SM58/57/94, etc.), electrónicos (Protools HD2, D-Command 24 faders, Mac Pro, Preamplificadores;Avalon Vt737Sp, Rupert Neve Portico 5012, Earthworks ZDT, Workstation Tascam SX-1, Interface Digidesign Digi002 Rack, Yamaha SPX990, Tascam CDRW700, unidad de reverberación análoga Orban, grabadores portátiles; Sound device 788T, Tascam HD-P2, Tascam DA-P1, Adat XT), software (Protools 8.0 HD, ProTool 8.0 LE, Samplitude 8.0, Smaart v.7.), y monitores (Genelec 8040, Yamaha NSM10, Yamaha HS80, Yamaha HS-10W).</div><br>
          <div style="text-align: justify; "><strong><em>Bibliotecas</em></strong></div><br>
          <div style="text-align: justify; ">La Universidad Austral de Chile cuenta con un Sistema de Bibliotecas destinado a las necesidades de información de todos los miembros de la comunidad universitaria: Biblioteca Central, Biblioteca de Medicina, Biblioteca Tecnológica de Ingeniería (ubicada en el Campus Miraflores donde se realizan las actividades del programa), Biblioteca de la Sede Puerto Montt y la Biblioteca Campus Patagonia. Dichas bibliotecas, además de ofrecer las modalidades tradicionales de préstamo de material, poseen servicios de biblioteca virtual, referencia electrónica, reserva anticipada, préstamo interbibliotecario y capacitación de los usuarios.</div><br>
          <div style="text-align: justify; "><strong>Biblioteca Tecnológica de Ingeniería (Edificio 12000):               </strong> Se posiciona como el centro neurálgico de la FCI, siendo un lugar de difusión de la información, un espacio interactivo entre personas, tecnología, documentos y un contenedor de conocimiento en sus distintos formatos. Además, otorga ambientes para el trabajo grupal situados al centro de la sala. Cuenta también con cubículos de estudio semi-traslúcidos en el área poniente y una sala protegida de la radiación solar para consultas virtuales, al lado oriente del edificio. Dispone de un espacio de 1000 m2, fue inaugurada en agosto de 2016 con una inversión total de &#36;816.794.382.</div><br>
           <div style="text-align: justify; "><strong>Biblioteca Central del Campus Isla Teja:           </strong> Cuenta con una superficie de 4.095 m2 y con capacidad para 611 usuarios. Además, dispone de cubículos individuales y grupales para estudio, investigación o preparación de trabajos y con tres pisos con acceso a conexión inalámbrica (WiFi). Se puede acceder además a una base de datos automatizada referida a la colección del sistema, de un catálogo en línea, sistema de estantería abierta, préstamo automatizado y servicios de biblioteca local y virtual. Dispone, además, entre sus instalaciones, más de 150 equipos computacionales para uso de los estudiantes ubicados en dos puntos del Campus Isla Teja, distribuidos en cinco laboratorios ubicados en el Edificio de Servicios Estudiantiles, construido frente al Edificio Nahmías, y en la Sala Universia, localizada en el Pabellón Docente.</div><br>

### Becas y evolución de la ayuda estudiantil
