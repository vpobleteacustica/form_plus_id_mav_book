El plan incrementar la tasa institucional de acreditación de magíster de 35% en 2015 a 77% en 2021. ## _Difusión del programa_

<table align="left" style="width:100%">
    <tbody style="font-size: 14px";>
        <tr>
            <td style="width: 7px;background:#036ED1;"></td>
            <td style="outline: thin solid" rowspan="1" colspan="1"><div style="text-align: justify; ">El instrumento base de difusión del programa es el sitio institucional <a href="http://www.acusticauach.cl">www.acusticauach.cl</a>, administrado desde el Instituto de Acústica, y cuya información se nutre de la descripción del programa en sus distintos componentes (denominación, objetivos, plan de estudios, líneas de investigación, cuerpo docente, infraestructura, periodo de postulaciones, requisitos de admisión, sistema de selección, costo de aranceles, becas disponibles, calendario académico, otros) oficialmente sancionada en las instancias institucionales respectivas. Desde esa información base se nutren, además, materiales de difusión complementarios como folletería, catálogos, recursos de difusión específicos para distribución en eventos internacionales, ferias de postgrado o afines. En todos los casos, al tratarse de instrumentos de difusión articulados sobre la base de su coherencia con respecto a las descripciones del programa institucionalmente sancionadas, incluidas sus actualizaciones, la información que se otorga a los estudiantes (y al público en general) es consistente y fidedigna respecto de las características reales del programa y de los compromisos de la institución en relación con los servicios ofrecidos y el respeto a las condiciones de formación declaradas en esas descripciones.</div><br>
