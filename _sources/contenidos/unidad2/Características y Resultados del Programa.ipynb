### Objetivos del programa 

<table align="left" style="width:100%">    
  <tbody style="font-size: 14px";>
  <tr style="outline: thin solid">
      <td style="width: 7px;background:#036ED1;"></td>
  <td bgcolor="white"><div style="text-align: justify; ">El MAV tiene como principal objetivo formar postgraduados especialistas, que cuenten con conocimiento avanzado en el campo de acústica y vibraciones*, mediante la entrega de conceptos teóricos, actividades prácticas de laboratorios, estudio individual guiado y el desarrollo de una tesis de magíster. La tesis debe tener un carácter de originalidad en el tema o en la metodología y capacitará al graduado del programa para el perfeccionamiento, para la innovación tecnológica y para el desarrollo de investigación científica con un alto grado de independencia.</div><br>

<div style="text-align: justify; ">El programa entrega una formación apropiada tanto como para proseguir estudios doctorales como para proyectar y desarrollar actividades profesionales avanzadas en el desarrollo científico y tecnológico en acústica y vibraciones.</div><br>
<div style="text-align: justify; "><sup style="font-size: 14px;">*</sup>El alcance disciplinar del programa se establece en la <em>acústica</em> como ciencia que estudia el sonido y <em>vibraciones</em> como fenómeno físico vinculado con el sonido.</div>

### Áreas o Líneas de investigación o creación, según corresponda

<table align="left" style="width:100%">    
  <tbody style="font-size: 14px";>
  <tr style="outline: thin solid">
      <td style="width: 7px;background:#036ED1;"></td>
  <td bgcolor="white"><div style="text-align: justify; ">El programa contempla dos líneas de investigación: i) acústica y ii) vibraciones.</div><br>

  <strong><em>Acústica:</em></strong> Esta línea estudia modelar, analizar y predecir el comportamiento de sistemas acústicos de alta complejidad, resolver problemas prácticos proponiendo soluciones innovadoras y sustentables, a partir de un proceso de investigación científica, desarrollada de acuerdo a los procedimientos y exigencias que imponen las técnicas modernas en acústica. Se incluyen en esta área las sublíneas de instrumentación y sensores acústicos, acústica forense, análisis de señales acústicas, acústica ambiental, acústica musical, comportamiento de materiales acústicos, modelamiento acústico, y acústica de salas.

<strong><em>Vibraciones:</em></strong> Esta línea estudia modelar, analizar y predecir el comportamiento de sistemas vibratorios de alta complejidad, resolver problemas prácticos proponiendo soluciones innovadoras y sustentables, a partir de un proceso de investigación científica, desarrollada de acuerdo a los procedimientos y exigencias que imponen las técnicas modernas en vibraciones. Se incluyen en esta área las sublíneas de instrumentación y sensores en vibraciones, análisis de señales vibratorias, modelamiento numérico de vibraciones, comportamiento y propiedades vibratorias de materiales y diseño mecánico-vibratorio de materiales.

## _Requisitos de admisión y proceso de selección_
### Requisitos de admisión

<table align="left" style="width:100%">    
  <tbody style="font-size: 14px";>
  <tr style="outline: thin solid">
      <td style="width: 7px;background:#036ED1;"></td>
  <td bgcolor="white"><div style="text-align: justify; ">El Reglamento General de Programas de Magíster (Anexo 3.1: Art. 19) establece, como requisito transversal para el ingreso a programas de magíster en la universidad, estar en posesión del grado académico de licenciado o de un título profesional cuyo nivel y contenidos sean equivalentes a los requeridos para obtención de una licenciatura. El Reglamento estipula además que, en el caso de postulantes extranjeros [no hispanohablantes], es requisito acreditar un dominio adecuado de español.</div><br>

<div style="text-align: justify; ">Con base en esas estipulaciones generales, los interesados en ingresar al Magíster en Acústica y Vibraciones deben cumplir con los siguientes requisitos específicos de admisión:</div>     
  <ol>
  <li><div style="text-align: justify; ">Estar en posesión del grado académico de licenciado o de un título profesional en el área de las ciencias de la ingeniería cuyo nivel y contenidos sean equivalentes a los requeridos para la licenciatura. </div></li>
  <li><div style="text-align: justify; ">En casos excepcionales, el programa podría considerar la evaluación de postulantes con grados académicos o títulos profesionales en áreas afines a la acústica y vibraciones en disciplinas distintas de la ingeniería. En dichos casos, los postulantes deben acreditar una formación robusta en matemáticas, física y fundamentos de la acústica. La acreditación de esa formación previa puede realizarse mediante la certificación de las calificaciones en cursos aprobados sobre esas materias del modo descrito en el requisito 3 o a través de la aprobación satisfactoria de un plan de nivelación diseñado por el Comité de Programa a postulantes preseleccionados.</div></li>
  <li><div style="text-align: justify; ">Acreditar un rendimiento académico sobresaliente o destacado en los estudios realizados que refleje concordancia con las exigencias de conocimientos previos definidas en las actividades curriculares del Plan de Estudios del programa. Para el cumplimiento de este requisito, los postulantes deben presentar los certificados de notas de los cursos rendidos en los estudios previos, presentados como antecedentes en la postulación, con las correspondientes calificaciones e información sobre el sistema de calificación, visados por la institución de estudios que los emite.</div></li>
  <li><div style="text-align: justify; ">Presentar un Currículum Vitae que documente la trayectoria de actividades académicas y profesionales realizadas pertinentes y de relevancia al interés de profundización de estudios en el programa. </div></li>
  <li><div style="text-align: justify; ">Adjuntar dos cartas de recomendación o informes de académicos que avalen la postulación a un programa de este tipo.</div></li>
</ol><br>

<div style="text-align: justify; ">Los requisitos de admisión del MAV y el mecanismo de presentación de los documentos se informan: a) en el sitio web del programa <a href="#">www.magister.acusticauach.cl;</a> b) en los instrumentos de difusión de la Escuela de Graduados de la Facultad de Ciencias de la Ingeniería en sus distintos soportes (folletería impresa, insertos, posters, otros); y c) en la «Plataforma de Postulaciones en Línea» implementada por la Dirección de Estudios de Postgrado <a href="#">www.uach.cl/postgrado.</a></div><br>
    
<div style="text-align: justify; ">Las postulaciones se formalizan y activan mediante Solicitud de Admisión*, a través de la «Plataforma de Postulaciones en Línea». El proceso de postulación se completa una vez que la Solicitud de Admisión se <em>envía</em> a través de esta plataforma acompañada de todos los documentos especificados en los requisitos: a) los certificados de cada uno de los títulos y grados obtenidos y declarados en la postulación; b) las concentraciones de notas respectivas (en ambos casos, en documentos oficializados por la institución que otorgó el grado o título); d) el currículum vitae con los antecedentes relevantes de interés para el programa; y d) las cartas de recomendación o informes de académicos que avalan la postulación.</div><br>

<div style="text-align: justify; "><sup style="font-size: 14px;">*</sup>El periodo comprendido entre el inicio de actividades del programa en 2008 y el primer semestre de 2017, la Solicitud de Admisión y la presentación de los documentos se realizaba mediante su entrega directa al Comité de Programa, en formato papel (en persona o vía correo ordinario) o en fotocopia digital (vía correo electrónico). Con la habilitación de la «Plataforma de Postulaciones en Línea» a partir del segundo de 2017, el sistema de postulaciones registró un mejoramiento significativo, tanto por el incremento en el volumen de potenciales postulantes como por la calidad del sistema de registro y trazabilidad de la información, que facilita su seguimiento, y permite la construcción de indicadores para orientar la toma de decisiones. La plataforma desarrollada incluye, además del sistema de postulaciones, el espectro completo de actividades administrativas y curriculares de los programas de postgrado, desde la admisión hasta el egreso y graduación de los estudiantes.</div><br>

### Articulación pregrado-postgrado

<table align="left" style="width:100%">    
  <tbody style="font-size: 14px";>
  <tr style="outline: thin solid">
      <td style="width: 7px;background:#036ED1;"></td>
  <td bgcolor="white"><div style="text-align: justify; ">El Art. 48 del Reglamento General de Magíster (Anexo 3.1) dispone que los estudiantes de pregrado de la Universidad Austral de Chile pueden postular a programas de magíster a través de programas de vinculación entre su carrera de pregrado y el postgrado. La forma de ingreso y estructura del currículo se describe en las normativas desarrolladas para este efecto por la Vicerrectoría Académica de la universidad y por acuerdos específicos de vinculación entre una escuela de pregrado y el programa de magíster. Estos "Acuerdos" establecen la duración del programa de vinculación, los cupos disponibles, los requisitos de admisión y las asignaturas o módulos que serán convalidados. En la sección 1.3 se documentó que el Magíster en Acústica y Vibraciones contempla un sistema de vinculación pregrado-postgrado con la carrera de Ingeniería Civil Acústica de la Facultad de Ciencias de la Ingeniería. Este sistema se ha articulado, primero, con arreglo a los Lineamientos para la Vinculación Pregrado-Postgrado de 2010 (Anexo 3.4); al Procedimiento de Vinculación entre Carreras de Pregrado y Programas de Magíster de 2013 (Anexo 3.5); y a la actualización de ese Procedimiento en 2017 (Anexo 3.6).
</div><br>

<div style="text-align: justify; ">El programa contó con un primer Acuerdo en 2009 que estuvo vigente hasta 2013 y un segundo Acuerdo suscrito en 2014 que se mantuvo vigente hasta 2020 (Anexo 10). La modalidad actualmente en vigencia en el programa es el Acuerdo de Vinculación de 2021 (Anexo 11) que mantiene las principales características del Acuerdo de 2014 e incorpora como innovación la gestión y trazabilidad en línea de las vinculaciones. El Acuerdo establece los siguientes términos para los estudiantes interesados en ingresar al MAV a través de este sistema:</div><br>
    <ol type="a" >
      <li style="text-align: justify; ">Duración de la vinculación: dos semestres como máximo.</li>
      <li style="text-align: justify; ">Cupos anuales disponibles: ocho.</li>
      <li style="text-align: justify; ">Requisitos de ingreso: haber cumplido con los requisitos para la obtención de grado de Licenciado en Ciencias de la Ingeniería. </li>
      <li style="text-align: justify; ">Contar con un promedio general acumulado igual o superior a 5,0 (en una escala de 1 a 7) en la carrera de Ingeniería Civil Acústica. </li>
    </ol>
    <div style="text-align: justify; ">La verificación de los requisitos y selección de los postulantes la realiza el Comité de Programa en conjunto con el Director de la Escuela de Ingeniería Civil Acústica quien participa como invitado. Los detalles del sistema de postulación, selección, aprobación e incorporación final al programa como alumnos regulares de magíster se informan en los Anexos 3.6 y 11.</div><br>
      <div style="text-align: justify; ">Los estudiantes aceptados en la articulación pueden convalidar hasta un máximo de tres asignaturas de su plan de estudios de pregrado (detalladas en el Anexo 11) además de una actividad equivalente al trabajo terminal de su titulación en pregrado. A partir de 2018, es obligatorio cursar la actividad ACUS 397: Seminario de Tesis como trabajo final de titulación. Las asignaturas susceptibles de incluir en el programa pueden ser cualquiera de las disponibles en la malla curricular del Magíster en Acústica y Vibraciones. El estudiante define las asignaturas de su programa de articulación con el Director de Programa y con el Director de la Escuela de Ingeniería Civil Acústica.</div><br>

### Estrategias pedagógicas y sistema de evaluación del desempeño académico de los estudiantes

<table align="left" style="width:100%">
  <tbody style="font-size: 14px";>
  <tr style="outline: thin solid">
      <td style="width: 7px;background:#036ED1;"></td>
   <td colspan="4"><div style="text-align: justify; ">Los estudiantes aceptados y matriculados en el programa deben cumplir con un plan de estudios de cuatro semestres académicos. Todas las asignaturas son semestrales y se dictan sólo una vez al año. El programa está organizado en la forma de régimen curricular semiflexible con asignaturas obligatorias y electivas. El currículo optativo y la tesis de grado permiten que el graduado integre y profundice los conocimientos básicos, según sea su interés personal, en alguna de las áreas principales del programa. Las materias ofrecidas dentro del currículo optativo constituyen una oportunidad para los estudiantes de adquirir conocimientos actualizados en temas más específicos.</div><br>

<div style="text-align: justify; ">Durante los primeros dos semestres deben completar un currículo obligatorio de trece créditos y un currículo electivo de, al menos, nueve créditos. A partir del tercer semestre (o antes, al considerar el Seminario de Tesis como comienzo del proceso), los estudiantes inician la Tesis de Grado (10 créditos), completando un total de 32 créditos. Reglamentariamente, el programa podría considerar la convalidación de cursos o créditos. Este mecanismo está restringido a una permanencia activa mínima de dos semestres en el programa y no considera la actividad de graduación.</div><br>

<div style="text-align: justify; ">Las asignaturas del magíster son teórico-prácticas y para la ejecución de éstas, se consideran clases formales en aula, seminarios, trabajos prácticos, presentaciones, salidas a terreno, estudio de casos, talleres, foros, clases interactivas o análisis de trabajos científicos. Los estudiantes tienen la posibilidad de hacer observaciones críticas constructivas para una mejor comprensión de los contenidos.</div><br>

<div style="text-align: justify; ">Durante el desarrollo del programa el estudiante puede reprobar solo una asignatura, siempre que en dicho semestre su promedio semestral ponderado (PSP) sea igual o mayor a 4,5. La asignatura cuya calificación sea inferior a 4,5 será reprobada y deberá repetirse en la primera oportunidad que se dicte y sólo por una vez. El plan de estudios quedará finalizado una vez que el estudiante haya aprobado todas sus asignaturas inscritas, y su promedio general ponderado (PGP) sea igual o superior a 5,0, habiendo acreditado también conocimientos de inglés mediante un examen de suficiencia en dicho idioma, lo que se calificará con la letra A (aprobado) o R (reprobado); en caso de reprobación el estudiante deberá cursar la asignatura de Idioma Inglés. Todo lo anterior, conducirá a que el estudiante adquiera la condición de candidato al grado de magíster.</div><br> 

<div style="text-align: justify; ">Junto al desarrollo de asignaturas lectivas, los estudiantes pueden comenzar a trabajar en el diseño de su proyecto de tesis durante el segundo semestre (Seminario de Tesis). El tema de tesis es seleccionado por el estudiante de acuerdo a sus intereses y a las líneas de investigación de los académicos.</div><br>

<div style="text-align: justify; ">El Proyecto de Tesis debe rendirse al final del primer año lectivo, momento en que se asigna una Comisión de Tesis, constituida por tres profesores del programa, incluido el Profesor Patrocinante. La Comisión realiza el seguimiento del estudiante durante esta segunda etapa con el fin de asegurar el buen cumplimiento del programa de estudios. Tanto el desarrollo de la fase experimental como la etapa de redacción y escritura de la Tesis de Grado, se deben realizar durante el tercer y cuatro semestres. Una vez finalizada la Tesis y entregado el documento escrito, ésta es evaluada por la Comisión de Tesis. Posteriormente, el estudiante debe aprobar un Examen de Grado, para concluir su formación, y obtener el grado de magíster.</div><br>

### <span style="background-color:lightgreen">Sistema de graduación</span> 

<table align="left" style="width:100%">
  <tbody style="font-size: 14px";>
  <tr style="outline: thin solid">
      <td style="width: 7px;background:#036ED1;"></td>
   <td colspan="1"><div style="text-align: justify; ">Las normas de graduación del Magíster en Acústica y Vibraciones se orientan en las estipulaciones establecidas en el Reglamento General de Magíster que dispone el siguiente sistema de graduación (Arts. 40-47, Anexo 3.1).</div><br>
       <div style="text-align: justify; "><strong>a) Requisitos de graduación (examen, defensa, publicación, otros)</strong></div><br>
       <div style="text-align: justify; ">El primer requisito que da inicio al proceso de graduación es haber completado satisfactoriamente el plan de asignaturas del programa. Es decir, haber aprobado todos los cursos del currículum obligatorio, los nueve créditos en electivos, la suficiencia de idioma inglés y la defensa del proyecto de investigación desarrollado en el Seminario de Tesis. El MAV define en sus objetivos el desarrollo de la tesis como la actividad terminal del programa. La tesis sólo puede ser dirigida por un académico acreditado para magíster y que pertenezca al claustro del programa. En la nomenclatura de la Universidad Austral de Chile al profesor guía se le denomina Profesor Patrocinante.</div><br>
        <div style="text-align: justify; ">Con la tesis evaluada, el estudiante está en condiciones de realizar la defensa de su trabajo de tesis en un Examen de Grado, oral y público.</div><br>
       <div style="text-align: justify; "><strong>b) Características de la tesis</strong></div><br>
       <div style="text-align: justify; ">El Reglamento General de Magíster define la tesis de magíster como un trabajo original individual de investigación, en un campo determinado del conocimiento relevante al programa, que permite demostrar que el estudiante ha adquirido los conocimientos, habilidades y aptitudes propias del programa. La Escuela de Graduados tiene la atribución de aprobar el reemplazo del formato de publicación tradicional de la tesis por un artículo enviado a una revista indexada (ISI, SCIELO o equivalente), a condición de que se trate de un trabajo elaborado durante el desarrollo del programa, como autor principal, con resultados del proyecto de investigación y aprobado por el profesor patrocinante.</div><br>
       <div style="text-align: justify; "><strong>c) Criterios de evaluación aplicados</strong></div><br>
       <div style="text-align: justify; ">Una vez finalizada la tesis, el estudiante debe entregar un ejemplar al Director de la Escuela de Graduados para proceder a su evaluación por la Comisión Evaluadora. Los trabajos deben estar escritos en castellano o inglés, con un resumen adicional en ambas lenguas. La Comisión Evaluadora está constituida por el profesor patrocinante y dos profesores informantes del programa. El Magíster en Acústica y Vibraciones se ha planteado como política que cada tesis de sus graduados cuente en su Examen de Grado con un profesor externo a la universidad experto en la disciplina o tema de la tesis. Como mecanismo para la implementación de esta política, el programa cuenta con un cuerpo de diecisiete profesores visitantes. El resultado: la totalidad de los veinticuatro graduados en los diez de funcionamiento del programa ha contado con un profesor externo en su comisión evaluadora.</div><br>
       <div style="text-align: justify; ">La tesis puede ser aprobada, aprobada con modificaciones o reprobada. La aprobación debe ser por unanimidad. Si fuese aprobada con modificaciones, el candidato debe entregar, visada por el profesor patrocinante, una versión corregida en un plazo no superior a treinta días, momento en el cual se procederá a la calificación. La calificación es en la escala de 1 a 7, aprobándose con una nota mínima de 5,0. Si fuese reprobada, la comisión autorizará por una sola vez la presentación de una nueva versión del trabajo en un plazo máximo de seis meses.</div><br> 
        <div style="text-align: justify; ">El Examen de Grado es el acto público y solemne en el cual el candidato debe exponer y defender la Tesis. El Examen de Grado se aprueba con el acuerdo unánime de los miembros de la Comisión Evaluadora. Se califica en una escala de 1 a 7, siendo la nota mínima de aprobación 5,0 (cinco coma cero). Si el Examen de Grado fuere reprobado, el candidato puede rendirlo por segunda vez en un plazo máximo de noventa días.</div><br>  
        <div style="text-align: justify; ">La nota final de graduación se obtiene ponderando en un 60% el promedio de notas de las asignaturas (PGP), 20% la nota de la tesis y 20% la nota del Examen de Grado. La nota final de graduación así calculada es equivalente a un concepto laudatorio de acuerdo con la siguiente escala:</div><br>

      <td bgcolor="white"><div style="text-align: justify; ">El programa cuenta con mecanismos formalizados de seguimiento académico de los estudiantes. De acuerdo con el Reglamento de Escuelas de Graduados y el Reglamento General de Magíster (Anexos 3.2 y 3.1), el seguimiento de la progresión académica en todas sus etapas recae de modo directo sobre el Comité de Programa, que tiene por función evaluar y supervisar la actividad del programa y entregar un informe anual sobre sus indicadores (estado de la matrícula, tasas de retención, tasas de graduación, tiempos de permanencia y otros) a la Dirección de la Escuela de Graduados de la Facultad.</div><br>
      <div style="text-align: justify; ">Para la ejecución de estas tareas, el Comité de Programa, en conjunto con la Dirección de la Escuela de Graduados, realiza reuniones periódicas (no inferiores a dos por semestre) para analizar la progresión de los estudiantes. En caso de indicadores negativos, se procura un levantamiento oportuno de la información e implementación de las opciones de solución disponibles.</div><br>
       <div style="text-align: justify; ">El seguimiento de la progresión académica se realiza también de modo indirecto a través de la Dirección de Estudios de Postgrado, mediante la solicitud periódica (de al menos una vez al año) de informes sobre resultados globales del programa a la Dirección de Escuela de Graduados a la que está adscrito el programa. El seguimiento de esta información se realiza en el contexto de los compromisos institucionales declarados en el eje estratégico de postgrado del Plan Estratégico Institucional (Anexo 12) y son monitoreados anualmente por la Dirección de Desarrollo Estratégico a través de sus tres unidades dependientes: i) la Unidad de Planificación Estratégica; ii) la Unidad de Calidad y Autoevaluación Institucional en el marco de los compromisos suscritos en la Acreditación Institucional vigente; y iii) la Unidad de Análisis Institucional en cumplimiento de los diversos requerimientos institucionales de información.</div><br>
     <div style="text-align: justify; "><strong>Acciones de mejoramiento</strong></div><br>
     <div style="text-align: justify; ">En el nivel institucional, la Dirección de Estudios de Postgrado, en un trabajo conjunto con el Consejo de Postgrado y la Dirección de Tecnologías de Información de la universidad, inició en 2015 el desarrollo e implementación de un nuevo Sistema de Gestión Administrativa y Curricular de Postgrado que ha permitido un mejoramiento significativo en el sistema de registro y trazabilidad de la información sobre la progresión académica de los estudiantes en todas sus etapas. El Sistema inició su puesta en marcha (en modalidad piloto) a partir del segundo semestre de 2017 y se encuentra plenamente operativo a partir del primer semestre de 2018.</div><br>
    <div style="text-align: justify; ">En el nivel específico del programa, como parte del proceso de autoevaluación realizado, se ha incluido en el Plan de Desarrollo para el periodo 2019-2027, la inclusión de las siguientes acciones de mejoramiento de la progresión de los estudiantes (el énfasis de las medidas es particularmente el mejoramiento de los tiempos de permanencia en el programa y específicamente los tiempos que dedican los estudiantes a la fase desarrollo de tesis, identificada como la principal debilidad del programa en este aspecto):</div><br>
<div style="text-align: justify; "><ol type="1">
  <li>Activar y potenciar los mecanismos previstos en el curso obligatorio Seminario de Tesis como instancia que garantice la elaboración, defensa e inscripción oportuna del Proyecto de Tesis.</li>
  <li>Incorporar y potenciar la modalidad de tesis en formato artículo, del modo en que lo contempla el Reglamento de General de Magíster (Anexo 3.1).</li>
  <li>Implementar una reunión semestral del claustro de programa en la que se presente el estado semestral de la progresión académica de los estudiantes. De modo específico, se solicitará a los académicos patrocinantes de tesistas que informen el estado de avance de las tesis en desarrollo (como instancia para identificar y comprometer propuestas de solución que agilicen el desarrollo y término de las tesis en curso).</li>
    <li>Potenciar el uso continuo de los espacios para estudiantes tesistas del programa recientemente implementados en el marco del Proyecto InnovING 2030 (v. sección 5.1.8 abajo).</li>
    <li>Implementar un sistema de becas de arancel específicamente de la Facultad de Ciencias de la Ingeniería dirigida a tesistas de los programas de magíster de la Facultad en etapa intermedia o avanzada a las que se acceda vía concurso interno en la Escuela de Graduados.</li>

### Mecanismo de seguimiento de graduados 

<table align="left" style="width:100%">
  <tbody style="font-size: 14px";>
  <tr style="outline: thin solid">
      <td style="width: 7px;background:#036ED1;"></td>
      <td bgcolor="white"><div style="text-align: justify; ">El Magíster en Acústica y Vibraciones dispone de tres tipos de mecanismos de seguimiento de sus graduados: uno en el nivel del programa, un segundo nivel a través de la Escuela de Graduados de la Facultad de Ciencias de la Ingeniería (FCI) y otro en el nivel institucional.</div><br>
      <div style="text-align: justify; ">El programa mantiene comunicación permanente con sus graduados y mantiene una base de datos actualizada y de acceso público de cada uno de ellos en la página web en <a href="http://www.acusticauach.cl">www.acusticauach.cl</a>.</div><br>
     <div style="text-align: justify; ">La Escuela de Graduados de la FCI cuenta con un sistema de seguimiento de egresados que consiste en mantener una base de datos actualizada con la información correspondiente a la inserción laboral de sus graduados además de otros antecedentes relevantes.</div><br>
    <div style="text-align: justify; ">En el nivel institucional, se creó en 2014 la Unidad de Vinculación con Egresados de la Universidad Austral de Chile (v. Anexo 14), dependiente de la Dirección de Vinculación con el Medio. Dentro de las funciones de esta unidad, está la gestión de las bases de datos con información actualizada de los egresados y la realización de estudios de seguimiento de egresados que apoyen la toma de decisiones a nivel central y de las unidades académicas. A partir de 2016, la unidad inició la aplicación de una encuesta anual a graduados de programas de postgrado UACh que busca aportar antecedentes para evaluar el impacto del postgrado en la trayectoria laboral y la satisfacción de los graduados respecto al proceso formativo. La unidad es responsable, asimismo, de coordinar la Red de Egresados de la Universidad Austral de Chile. Esta Red tiene por tarea generar acciones que promuevan el vínculo activo y continuo con ex alumnos.</div><br></td>
