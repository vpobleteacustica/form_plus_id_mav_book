## Definiciones reglamentarias

### Modalidades utilizadas y criterios considerados para la selección e incorporación de académicos al programa

<table align="left" style="width:100%">
    <tbody style="font-size: 14px";>
        <tr>
            <td style="width: 7px;background:#036ED1;"></td>
            <td style="outline: thin solid" rowspan="1" colspan="1"><div style="text-align: justify; ">El                     programa de Magíster en Acústica y Vibraciones dispone de un conjunto de instrumentos                         formalmente  implementados que establecen los mecanismos, fijan los criterios y orientan las                 acciones para la selección e incorporación de académicos al programa (en las modalidades de                   claustro, colaborador o visitante) a través del sistema institucional de acreditación de profesores de magíster (se detalla en los párrafos siguientes). Este sistema permite habilitar e incorporar, en los programas que así lo solicitan, a profesores con o sin vínculo contractual con la universidad. El sistema se complementa, en consecuencia, si bien de modo indirecto, con la normativa institucional para la contratación, adscripción o promoción de docentes en las plantas ‘académica’, ‘adjunta’ o de ‘profesores visitantes’ de la institución.</div><br>
            <div style="text-align: justify; ">La selección de los académicos susceptibles de integrar el cuerpo docente del Magíster en Acústica y Vibraciones, con independencia de si se trata de profesores externos o de académicos de planta, se realiza preliminarmente en la Dirección de la Escuela de Graduados de la Facultad de Ciencias de la Ingeniería, a propuesta del Comité de Programa. Las propuestas del Comité de Programa se articulan principalmente en función del fortalecimiento de las líneas de investigación o actualizaciones del plan de estudios. En suma, de potenciar aspectos del desarrollo del programa identificados como susceptibles de mejoramiento en los procesos de autoevaluación. Una política institucional es que dichas propuestas se encuentren, a su vez, en sintonía con el Plan de Desarrollo de Postgrado de la Facultad o Sede a la que está adscrito el programa —en este caso, el Plan de Postgrado de la Facultad de Ciencias de la Ingeniería (Anexo 13)— y de los compromisos de la universidad en sentido amplio en relación con el postgrado declarados en el Plan Estratégico Institucional en vigencia (Anexo 12).</div><br>
            <div style="text-align: justify; ">Los académicos, por tanto, primero son propuestos por el Comité de Programa a la Dirección de la Escuela de Graduados de la Facultad. Una vez que esta instancia otorga su respaldo a la solicitud, los docentes deben postular para ser ‘acreditados’ o ‘autorizados’ por la Comisión Central de Magíster, mediante la presentación de sus antecedentes a esa instancia a través de la Dirección de Estudios de Postgrado. Este procedimiento de acreditación interna se encuentra normado a través del Decreto de Rectoría 024-2014 que establece los Criterios para la Acreditación de Profesores de Magíster (Anexo 3.7).</div><br>
            <div style="text-align: justify; ">Los Criterios de Acreditación difieren si se postula a un magíster académico o a uno de orientación profesional. Para el caso de los programas de carácter académico, como el MAV, los Criterios estipulan cuatro tipos de requisitos que deben cumplir los académicos para obtener la ‘acreditación’:</div>
            <ol>
                 <li><div style="text-align: justify; "><strong>Grado académico:</strong> Los docentes postulantes deben poseer, al menos, el grado de magíster en el área correspondiente. Excepcionalmente pueden ser acreditados aquellos que, no teniendo el grado de magíster, reúnen los demás requisitos y posean una trayectoria académica relevante o competencias profesionales destacadas.</div></li>
                 <li><div style="text-align: justify; "><strong>Productividad científica:</strong> Demostrable a través de: a) participación en una línea de investigación productiva, actualizada y sustentada en proyectos financiados con recursos internos o externos; b) publicaciones en revistas indexadas en ISI, Scopus, SCIELO o con comité editorial, libros o capítulos de libros y otros tipos publicaciones o modalidades de expresión académica de relevancia en el área; y c) registro de patentes, aceptadas o en proceso formal de trámite.</div></li>
                <li><div style="text-align: justify; "><strong>Docencia y formación de estudiantes:</strong>  Experiencia en docencia demostrable. Los criterios estipulan que, en el caso de no demostrar experiencia en docencia, los candidatos deben contar formalmente con habilitación pedagógica (exigida con carácter de obligatoria para todos los académicos a partir de 2017). Se valora asimismo la participación en patrocinio de tesis, trabajos de graduación o titulación u otros mecanismos de formación de estudiantes de pregrado.</div></li>
                <li><div style="text-align: justify; "><strong>Reconocimiento nacional e internacional: </strong> Se valora la participación en actividades relevantes en temas de la especialidad de alcance nacional o internacional. En los Criterios se indica de modo explícito que este aspecto cobra especial relevancia en la evaluación de aquellos postulantes que cuentan con una trayectoria profesional destacada y se los estima un aporte sustancial para los programas que soliciten su participación.</div></li>
            </ol>
            <div style="text-align: justify; ">El proceso de postulación a la acreditación se realiza a través del Sistema de Información Académica Administrativa (SIACAD), una herramienta online de la intranet corporativa a disposición de los docentes para la gestión de su información curricular. A través de esta herramienta, los docentes informan los antecedentes actualizados de su trayectoria en un formulario de currículum específico para acreditación de magíster académico en función de los requisitos requeridos. La postulación se remite, vía SIACAD, al Director de la Escuela de Graduados quien toma conocimiento de los antecedentes del postulante e incorpora sus observaciones en respaldo a la postulación y el tipo de actividades específicas para la cual se solicita la acreditación o autorización. Declaradas estas observaciones por parte de la Dirección de la Escuela de Graduados, los antecedentes quedan a disposición del Director de Estudios de Postgrado, quien los expone a la Comisión Central de Magíster para su revisión y dictamen.</div><br>
            <div style="text-align: justify; ">Una vez revisados los antecedentes, la Comisión resuelve el estatus de acreditación o autorización del postulante conforme a los siguientes criterios:</div>
            <ul>
                 <li><div style="text-align: justify; ">Los académicos que cumplen con los criterios de acreditación quedan habilitados para participar en todo tipo de actividades en programas de magíster. La acreditación los habilita de modo exclusivo para formar parte del claustro, dirigir tesis y ser responsables de asignaturas de nivel 300 [esto es, asignaturas que en la nomenclatura interna de la institución corresponden a cursos regulares para programas de magíster].</div></li>
                 <li><div style="text-align: justify; ">Los académicos que cumplen parcialmente con los requisitos pueden ser autorizados para colaborar en actividades acotadas en los programas, por ejemplo, participar como profesores colaboradores de asignaturas. Sin embargo, no pueden dirigir tesis ni formar parte del Comité de Programa. En casos excepcionales y debidamente justificados por la Escuela de Graduados que así lo solicita, la Comisión Central de Magíster puede extender la autorización a la responsabilidad de asignaturas.</div></li>
                <li><div style="text-align: justify; ">Los académicos o investigadores externos a la universidad cuyos antecedentes sean equivalentes a los exigidos a los profesores de la UACh, pueden igualmente ser acreditados como profesores de magíster.</div></li>
            </ul>
            <div style="text-align: justify; ">Los Criterios de Acreditación estipulan que la Comisión debe emitir una resolución fundada; que las acreditaciones tengan una validez máxima de cinco años (en virtud del mérito de sus antecedentes); y que, una vez vencidas, se solicite su renovación.</div><br>
            <div style="text-align: justify; ">El procedimiento de contratación se establece en el Art. 14 del Reglamento de Carrera Académica (Anexo 15). La Unidad Básica (Instituto) solicita abrir un concurso para llenar un cargo Académico. Aprobada esta solicitud, se envía con los requisitos y el perfil del candidato a la Vicerrectoría Académica con la finalidad de que lo autorice. Una vez autorizado, se publica el llamado a concurso en un diario de circulación nacional. La Dirección de Personal recibe los antecedentes de los postulantes y posteriormente los remite a la Facultad respectiva. El Consejo de Facultad selecciona una terna en base a la propuesta efectuada por el Claustro de la Unidad Académica Básica, la cual se remite a la Vicerrectoría Académica desde donde se presentan los antecedentes y la categoría académica propuesta para su contratación a la Comisión Central de Adscripciones y Promociones integrada por siete profesores titulares de la universidad que representan las distintas áreas del conocimiento.</div><br>
            </td>
        </tr>

### Procedimiento utilizado para la evaluación del desempeño docente 

<table align="left" style="width:100%">
    <tbody style="font-size: 14px";>
        <tr>
            <td style="width: 7px;background:#036ED1;"></td>
            <td style="outline: thin solid" rowspan="1" colspan="1"><div style="text-align: justify; ">La evaluación del desempeño docente de los académicos que participan en el programa se realiza a través de los siguientes mecanismos:</div>
            <ol type="a">
                <li><div style="text-align: justify; ">Evaluación del desempeño docente por parte de los estudiantes. A partir de 2013, la universidad implementó un sistema de evaluación del desempeño de los docentes en el contexto de las asignaturas que se imparten en los programas de doctorado y de magíster. El instrumento consiste en una encuesta en línea que responden los estudiantes cada semestre al término de cada curso.</div></li><br>
                <li><div style="text-align: justify; ">Evaluación institucional a través de convenios de desempeño. A nivel institucional, la evaluación de los docentes se desarrolla a través de la suscripción de convenios de desempeño individuales, acordados en el marco de las orientaciones de la Facultad (en este caso, la de Ciencias de la Ingeniería) y de la Universidad Austral de Chile (Reglamento sobre Compromisos de Desempeño, Anexo 10). La suscripción del convenio de desempeño es requisito para la promoción académica, la que se ciñe al Reglamento de Carrera Académica aprobado en su última versión en enero de 2013 (Anexo 10). Estos tienen una duración de tres años con seguimiento y control periódico anual, y evaluación al finalizar el período.</div></li><br>
                <li><div style="text-align: justify; ">Evaluación del desempeño académico en función de los procesos de acreditación. En el marco del proceso de autoevaluación del programa se ha desarrollado una encuesta que aborda aspectos similares a los descritos y que se aplica a estudiantes como a docentes. Los resultados de la encuesta a docentes realizada en el marco del proceso de autoevaluación informado en este Formulario se analizan en el Informe de Autoevaluación y pueden consultarse en el Anexo 23). Dichas encuestas se revisan, en primera instancia, en el Comité de Autoevaluación del programa (Anexo 21) para detectar áreas susceptibles de mejorar o intervenir. Los resultados deben presentarse además al Consejo de Escuela de Graduados de la Facultad, según lo dispone el Art. 12 letra e) del Reglamento de Escuelas de Graduados (Anexo 3.2).</div></li><br>
